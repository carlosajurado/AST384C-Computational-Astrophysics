```python
%matplotlib inline
import numpy
import matplotlib 
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline as US
import scipy.integrate as s_int
import time
```

```python
def f(x,t):
    return -x**3+numpy.sin(t)
```

```python
# start, end of interval:
a=0.0
b=10.0
```

```python
def euler_method(N):
    # get step width based on endpoints and desired number of intervals N
    h=(b-a)/N
    # get an array of N steps in time t from a to b. To include both endpoints, need N+1 values
    tpoints = numpy.linspace(a, b, N+1, endpoint=True)
    # set up an array of x values (initialized to 0) corresponding to t values
    xpoints = numpy.zeros(tpoints.size)
    count=0
    # xx will be my x variable in this function, initialized at x(0)=0
    xx=0
    for i in range(tpoints.size):
        xpoints[i]=xx
        xx += h*f(xpoints[i], tpoints[i])

    return(tpoints, xpoints)
```

```python
# lets compare results for a variety of values of N:
n1=10
n2=30
n3=100
n4=1000

euler1=euler_method(n1)
euler2=euler_method(n2)
euler3=euler_method(n3)
euler4=euler_method(n4)
```

```python
plot(euler1[0], euler1[1], 'k-', lw=2, label=r'$N={0}$'.format(n1))
plot(euler2[0], euler2[1], 'c-', lw=2, label=r'$N={0}$'.format(n2))
plot(euler3[0], euler3[1], 'm--', lw=2, label=r'$N={0}$'.format(n3))
plot(euler4[0], euler4[1], color='0.5', ls=':', lw=3, label=r'$N={0}$'.format(n4))
xlabel(r'$t$')
ylabel(r'$x ( t )$')
legend(frameon=False)
```
